# --- Variable Definitions ---
"""
CLIMATE & LOCATION VARIABLES
- country: Country in West Africa where the observation was recorded
- location_name: Specific city or locality within the country
- latitude, longitude: Geographic coordinates of the observation
- timezone: Time zone of the location
- last_updated_epoch: Unix epoch timestamp of the latest data update
- last_updated: Human-readable date/time of the last update (local time)

TEMPERATURE VARIABLES
- temperature_celsius: Air temperature in degrees Celsius (°C)
- temperature_fahrenheit: Air temperature in degrees Fahrenheit (°F)
- feels_like_celsius: Apparent temperature in °C considering humidity and wind
- feels_like_fahrenheit: Apparent temperature in °F

WIND & PRESSURE VARIABLES
- wind_mph: Wind speed in miles per hour
- wind_kph: Wind speed in kilometers per hour
- wind_degree: Wind direction in degrees (0° = North)
- wind_direction: Compass representation of wind direction (e.g., N, NE, SW)
- pressure_mb: Atmospheric pressure in millibars (hPa)
- pressure_in: Atmospheric pressure in inches of mercury (inHg)

PRECIPITATION & HUMIDITY
- precip_mm: Precipitation amount in millimeters
- precip_in: Precipitation amount in inches
- humidity: Relative humidity in percentage (%)
- cloud: Cloud cover percentage (%)

VISIBILITY & UV
- visibility_km: Visibility distance in kilometers
- visibility_miles: Visibility distance in miles
- uv_index: Ultraviolet radiation index
- gust_mph, gust_kph: Peak wind gust speeds in mph and kph

AIR QUALITY VARIABLES (verified WHO / EPA standards)
- air_quality_Carbon_Monoxide: CO concentration (µg/m³)
- air_quality_Ozone: O₃ concentration (µg/m³)
- air_quality_Nitrogen_dioxide: NO₂ concentration (µg/m³)
- air_quality_Sulphur_dioxide: SO₂ concentration (µg/m³)
- air_quality_PM2.5: Fine particulate matter (≤2.5 µm, µg/m³)
- air_quality_PM10: Coarse particulate matter (≤10 µm, µg/m³)
- air_quality_us-epa-index: U.S. EPA air quality index (1–6 scale)
- air_quality_gb-defra-index: UK DEFRA air quality index (1–10 scale)

ASTRONOMICAL DATA
- sunrise, sunset: Local times for sunrise and sunset
- moonrise, moonset: Local times for moonrise and moonset
- moon_phase: Lunar phase (e.g., New Moon, Waxing Gibbous)
- moon_illumination: Percentage of moon illumination (%)

DERIVED TEXT VARIABLES
- condition_text: Descriptive weather condition (e.g., Clear, Rain, Thunderstorm)
"""

```python
# Missing values summary ---
missing_summary = df.isnull().sum().sort_values(ascending=False)
missing_summary = missing_summary[missing_summary > 0]

print("Columns with Missing Values:\n")
display(missing_summary)

# cardinality check
print("\nColumns with high cardinality (>100 unique values):")
for col in df.columns:
    if df[col].nunique() > 100:
        print(f"{col}: {df[col].nunique()} unique values")
```

```python
# --- Numeric summary ---
num_summary = df.describe().T
#num_summary.to_csv("numeric_summary.csv")
num_summary.head(10)
```

```python
# Check duplicates
duplicates = df.duplicated().sum()
print(f"Duplicate rows: {duplicates}")


df['last_updated'] = pd.to_datetime(df['last_updated'], errors='coerce')


print("Date range:", df['last_updated'].min(), "to", df['last_updated'].max())


invalid_temp = df[df['temperature_celsius'] < -40]
print(f"Invalid low temperature records (< -40°C): {len(invalid_temp)}")

invalid_humidity = df[(df['humidity'] < 0) | (df['humidity'] > 100)]
print(f"Invalid humidity values: {len(invalid_humidity)}")

invalid_uv = df[df['uv_index'] < 0]
print(f"Invalid UV index values: {len(invalid_uv)}")


if not missing_summary.empty:
    corr_missing = df.isnull().corr()
    print("\nMissingness correlation matrix (if applicable):")
    display(corr_missing)
else:
    print("\nNo missingness detected.")
```
